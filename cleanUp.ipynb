```python
import pandas as pd

path = r'/Users/bennett/Documents/marathonEnvironment/base2019-2022.csv'

df = pd.read_csv(path, low_memory=False)
```

```python
df = df.rename_axis('ID')
df = df.drop(columns=['Unnamed: 0', 'Unnamed: 0.1', 'City__State', 'State__Country', 'Country', 'City', 'Address', 'Date_time', 'Resolved_Address', 'racename',
       'City__Country', 'int64_field_0', 'State'])

print(df.columns.tolist)
print(df.head)
```

```python
df['IsFrom'] = df['IsFrom'].astype(str)
df['broadest'] = df['IsFrom'].apply(lambda x: x.rsplit(',', 1)[-1].strip())
```

```python
df_distinct = df[['broadest', 'IsFrom', 'Race', 'Name', 'Location']]

df_distinct = df_distinct.drop_duplicates(subset=["IsFrom"], keep="first", inplace=False)

print(df_distinct)

outpath = r'/Users/bennett/Documents/marathonEnvironment/analysis/locationLearningSet.csv'

df_distinct.to_csv(outpath)
```
